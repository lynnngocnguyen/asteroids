```python
%matplotlib inline
import matplotlib.pylab as plt

import numpy as np
import pynbody as pb
import astropy.units as u

import KeplerOrbit as ko
```

```python
times, max_mass, mean_mass, n_particles, stepnumber = np.loadtxt('time_mass_asteroids_1.txt')
max_masss = (max_mass*u.M_sun).to(u.g)


mean_masss = (mean_mass*u.M_sun).to(u.g)

#do same for mean mass
```

```python


plt.loglog(stepnumber, max_masss, 'b', label='Max Mass')
plt.loglog(stepnumber, mean_masss, 'g',label='Mean Mass')
plt.xlabel('Time (Yr) ')
plt.ylabel('Mass (Grams)')
#plt.title('Time Vs. Masses')
plt.legend()
```
